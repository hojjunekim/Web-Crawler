```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import time
from selenium.webdriver.common.keys import Keys
import requests
import csv
import openpyxl
```

```python

try:
    wb = openpyxl.load_workbook("navernews.xlsx")
    sheet = wb.active
    print("불러오기 완료")
except:
    wb = openpyxl.Workbook()
    sheet = wb.active



def get_information(name_of_place):

    def getting_time_url(places):
        info = {}
        chromedriver = '/usr/local/bin/chromedriver'
        driver = webdriver.Chrome(chromedriver)
        driver.get('https://www.naver.com')
        time.sleep(1)
        search = driver.find_element_by_xpath('//*[@id="query"]')
        search_key = '+' + places[0] + ' +서울'
        search.send_keys(search_key)
        search.send_keys(Keys.ENTER)
        time.sleep(1)
        blog = driver.find_element_by_xpath('//*[@id="lnb"]/div/div[1]/ul/li[2]/a/span')
        blog.click()
        time.sleep(1)
        option = driver.find_element_by_xpath('//*[@id="_search_option_btn"]')
        option.click()
        time.sleep(0.5)
        time_range = driver.find_element_by_xpath('//*[@id="snb"]/div/ul/li[2]/a/span')
        time_range.click()
        whole = driver.find_element_by_xpath('//*[@id="_nx_option_date"]/ul/li[1]/a')
        whole.click()
        temp = []
        temp.append(driver.current_url)
        blogs = driver.find_element_by_xpath('//*[@id="main_pack"]/div[2]/div/span')
        temp.append(blogs.text.split(' ')[2][:-1])
        info[places[0]] = temp

        for place in places[1:]:
            time.sleep(0.5)
            search = driver.find_element_by_xpath('//*[@id="nx_query"]')
            search.clear()
            search_key = '+' + place + ' +서울'
            search.send_keys(search_key)
            search.send_keys(Keys.ENTER)
            time.sleep(0.5)
            time_range = driver.find_element_by_xpath('//*[@id="snb"]/div/ul/li[2]/a/span')
            time_range.click()
            whole = driver.find_element_by_xpath('//*[@id="_nx_option_date"]/ul/li[1]/a')
            whole.click()
            time.sleep(0.2)
            temp = []
            temp.append(driver.current_url)
            blogs = driver.find_element_by_xpath('//*[@id="main_pack"]/div[2]/div/span')
            temp.append(blogs.text.split(' ')[2][:-1])
            info[place] = temp

        return info


    def change_date(url, date):
        temp = url.replace('20030520', date)
        temp = temp.replace('option=0', 'option=8')
        last = 'from' + date + 'to20200427'
        new_url = temp[:-3] + last
        return new_url

    def review_by_time_range(information):
        final = {}
        for place_set in information.items():
            num = [place_set[1][1]]
            temp_url = []
            origin_url = place_set[1][0]
            temp_url.append(change_date(origin_url, '20200127'))
            temp_url.append(change_date(origin_url, '20191027'))
            temp_url.append(change_date(origin_url, '20190727'))
            temp_url.append(change_date(origin_url, '20190427'))
            temp_url.append(change_date(origin_url, '20190127'))
            temp_url.append(change_date(origin_url, '20181027'))
            temp_url.append(change_date(origin_url, '20180727'))
            temp_url.append(change_date(origin_url, '20180427'))
            temp_url.append(change_date(origin_url, '20180127'))
            temp_url.append(change_date(origin_url, '20171027'))
            temp_url.append(change_date(origin_url, '20170727'))
            temp_url.append(change_date(origin_url, '20170427'))


            for one_url in temp_url:
                response = requests.get(one_url)
                soup = bs(response.content, 'html.parser')
                try:
                    num.append(soup.find('span', {'class': 'title_num'}).text.split(' ')[2][:-1])

                except:
                    num.append(0)
            final[place_set[0]] = num
            sheet.append(num)

        return final

    testing_info= getting_time_url(name_of_place)
    final_result= review_by_time_range(testing_info)

    return final_result

test=  ['파크하얏트', '그랜드하얏트', 'JW메리어트','르메르디앙', '콘래드', '반얀트리', '인터컨티넨탈', '쉐라톤', '힐튼', '블루보틀']

result = get_information(test)
print(result)


```
